```python
import pandas as pd
import re
from cleantext import clean #https://www.educative.io/answers/how-to-remove-emoji-from-the-text-in-python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
# Load file
# Slang dictionary for normalization
slang_dict = pd.read_csv(r"/content/drive/MyDrive/metopen/new_kamusalay.csv", encoding='latin-1', header=None)
slang_dict = slang_dict.rename(columns={0: 'original',
                                      1: 'replacement'})

# HateComment Dataset
df = pd.read_csv (r"/content/drive/MyDrive/TFD Prep/HateComments Dataset - Comments.csv", encoding='latin-1')
df
```

```python
# Cleaning Data
def lowerCase(comment):
  return comment.lower()

# Remove emoticon or emojis from comment
def clean_emoticon(comment):
  comment = clean(comment, no_emoji=True)
  return comment

# Remove another unacesassary character on comment
def remove_unnecessaryChar(comment):
  #remove mention text
  comment = re.sub("@[A-Za-z0-9]+", ' ', comment)
  #remove non alphanumeric
  comment = re.sub('[\W_]+', ' ', comment)
  #remove multiple spaces
  comment = re.sub(r'\s+', ' ', comment)
  return comment

# Normalize slangs and replace it with original word in slang dictionary
slang_dict_map = dict(zip(slang_dict['original'], slang_dict['replacement']))
def normalization(comment):
  comment = ' '.join([slang_dict_map[word] if word in slang_dict_map else word for word in comment.split(' ')])
  return comment

# Stopword removing using sastrawi
id_stopword_dict = pd.read_csv('/content/drive/MyDrive/metopen/stopwordbahasa.csv', header=None)
id_stopword_dict = id_stopword_dict.rename(columns={0: 'stopword'})
def remove_stopwords(comment):
  comment = ' '.join(['' if word in id_stopword_dict.stopword.values else word for word in comment.split(' ')])
  comment = re.sub('  +', ' ', comment) # Remove extra spaces
  comment = comment.strip()
  return comment

# factory = StemmerFactory()
# stemmer = factory.create_stemmer()
# def stemming(comment):
#   return stemmer(comment)
```

```python
# Applying preprocessing to the comments
def preprocess(text):
  text = lowerCase(text)
  print(text)
  # comment = non_alphanumeric(comment)
  text = remove_unnecessaryChar(text)
  print(text)
  text = normalization(text)
  print(text)
  text = remove_stopwords(text)
# comment = stemming(comment)
  return text
```

```python
# uji coba penghapusan non alpha numeric
text = "ya Bani taplak/ kawan kawan @mfthl ya.   jgn begitu 😅😅 wkwkwk"
text_cleaned = preprocessing(text)
print(text_cleaned)
```

```python
# hapus white space 1x 2x 3x dan seterusnya
def remove_space (text):
    text = re.sub(r'\s+', ' ', text)
    return text

text_space = "This     is    an   example    with                                             multiple spaces."
try_space = remove_space(text_space)
print (try_space)
print (text_space)
```

```python
# excute penghpausan white space
df ['Original Text'] = df ['Original Text'].apply(preprocess)
df
```
